```python
# Load the raw data from Phase 2
DATA_PATH = '../data/raw/groceries.csv'
df = pd.read_csv(DATA_PATH)

# Clean Data: Remove the duplicates identified in Phase 2
before = len(df)
df_clean = df.drop_duplicates().copy()
after = len(df_clean)

print(f"Data Cleaning: Removed {before - after} duplicate rows.")
print(f"Current Shape: {df_clean.shape}")
```

```python
# Convert Date to datetime objects
df_clean['Date'] = pd.to_datetime(df_clean['Date'], format='%d-%m-%Y')

# Feature Engineering: Create 'Is_Weekend' (Saturday=5, Sunday=6)
df_clean['Is_Weekend'] = df_clean['Date'].dt.dayofweek.isin([5, 6])

print("Feature Engineering: Created 'Is_Weekend' column.")
df_clean.head()
```

```python
from src.scraping import scrape_food_categories

# 1. Get categories from Wikipedia
scraped_categories = scrape_food_categories()

# 2. Simple mapping function (You can expand this list)
def get_category(item):
    item = item.lower()
    if any(x in item for x in ['milk', 'yogurt', 'cheese', 'cream']): return 'Dairy'
    if any(x in item for x in ['veg', 'fruit', 'root', 'onion']): return 'Produce'
    if any(x in item for x in ['beer', 'wine', 'soda', 'water']): return 'Beverages'
    if any(x in item for x in ['bread', 'roll', 'bun', 'pastry']): return 'Bakery'
    if any(x in item for x in ['sausage', 'beef', 'pork', 'chicken']): return 'Meat'
    return 'Other'

# 3. Apply the integration
df_clean['Category'] = df_clean['itemDescription'].apply(get_category)

print("Data Integration: Wikipedia categories merged successfully.")
```

```python
# Final Formatting: Reorder for clarity
df_final = df_clean[['Member_number', 'Date', 'Is_Weekend', 'itemDescription', 'Category']]

# Save the prepared data
OUTPUT_PATH = '../data/processed/prepared_groceries.csv'
df_final.to_csv(OUTPUT_PATH, index=False)

print(f"FINAL PREPARED DATASET SUMMARY")
print(f"Shape: {df_final.shape}")
print(f"Prepared dataset saved to: {OUTPUT_PATH}")
df_final.head()
```
